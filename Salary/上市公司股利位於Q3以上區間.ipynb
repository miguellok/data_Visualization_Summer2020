```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_excel('上市公司.xlsx',skiprows=0)
data
```

```python
data.describe()
```

```python
bonus1=data["現金股利"]>2.2
w1=data['非擔任主管職務之全時員工薪資平均數(仟元/人)']>=657
d1=data[bonus1&w1]
d1
```

```python
bonus1=data["現金股利"]>2.2
p1=data['非擔任主管職務之全時員工薪資平均數(仟元/人)']<657
d2=data[bonus1&p1]
d2
```

```python
d1.groupby('產業別').count()
d1.groupby('產業別').count().to_csv("產業別1.csv")
d2.groupby('產業別').count()
d2.groupby('產業別').count().to_csv("產業別2.csv")
```

```python
data1=pd.read_csv('產業別1.csv',skiprows=0)
data1[['產業別','上市櫃狀態']]
```

```python
data2=pd.read_csv('產業別2.csv',skiprows=0)
data2[['產業別','上市櫃狀態']]
```

```python
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.figure(figsize=(9,9))

plt.plot(data1['公司名稱'],data1['產業別'],'^k',color='b',label='薪資高於平均數')
plt.plot(data2['公司名稱'],data2['產業別'],'^k',color='r',label='薪資低於平均數') 

plt.legend(fontsize=15)

plt.grid()
plt.xlabel('公司數總計',fontsize=15)
plt.ylabel('產業類別',fontsize=15)
plt.xticks(range(26),fontsize=15)
plt.yticks(fontsize=16)
plt.title('股利高於2.2(以產業別分類的公司數總計)',fontsize=15)
```
