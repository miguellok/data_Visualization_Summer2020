```python
lyrics_words = list(lyrics)
```

```python
lyrics_words
```

```python
result_list = []
for w in lyrics_words:
    if w != ' ' and w != '\n':
        result_list.append(w)
```

```python
len(result_list)
```

```python
result_list
```

```python
d = {}
for word in result_list:
    if word not in d:
        d[word] = 1
    elif word in d:
        d[word] = d[word]+1
```

```python
d.items()
```

## sort 是應用在 list 上的方法，sorted 可以對所有可迭代的對象進行排序操作。

```python
d_sorted = sorted(d.items(), key=lambda x: x[1], reverse=True)
```
