```python
# 把不要的列去掉
df_2330_Sep = df_2330_Sep.iloc[0:19]
df_2330_Oct = df_2330_Oct[:22]
```

```python
# 把兩個月的資料連接起來
df_processed = pd.concat([df_2330_Sep,df_2330_Oct],axis=0)
#df_processed.index = range(df_processed.shape[0])
# 把
#df_processed["成交股數"] = df_processed["成交股數"].apply(lambda x: x.replace(',',''))
df_processed["成交股數"] = pd.to_numeric(df_processed["成交股數"].apply(lambda x: x.replace(',','')))
```

```python
# 只取需要的column
df_processed = df_processed[["日期","開盤價","最高價","最低價","收盤價","成交股數"]]
```

```python
# 把價差算出來
df_processed["價差"] = df_processed["收盤價"] - df_processed["開盤價"]
#df_processed["價差"].sum()
#df_processed["價差"].plot()
df_processed.to_csv('2330_SepOct_processed.csv', sep=',',encoding = "cp950")
```
