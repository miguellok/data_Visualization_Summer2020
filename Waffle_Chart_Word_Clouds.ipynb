```python
import pandas as pd
import matplotlib.pyplot as plt
from pywaffle import Waffle
```

```python
df = {'Country':['Denmark (丹麥)','Norway (挪威)','Sweden (瑞典)'],'Num':[12,7,6]}
```

```python
country_dic = pd.DataFrame(df)
```

```python
country_dic
```

```python
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
fig = plt.figure(
    figsize=(8,6),
    FigureClass = Waffle,
    rows = 3,
    values = country_dic.Num,
    labels = list(country_dic.Country)
)
```

## Advantages:
* More visually attractive.
* Used for attractive Dashboards.

## Disadvantages:
* Waffle Chart becomes too complicated when too many segments are involved.
* individualized spaces do not leave a spot to put numbers or text within the visual itself.
